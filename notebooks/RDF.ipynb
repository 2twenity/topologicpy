```python
# Make sure you have pip installed rdflib

from topologicpy.Vertex import Vertex
from topologicpy.Edge import Edge
from topologicpy.Topology import Topology
from topologicpy.CellComplex import CellComplex
from topologicpy.Graph import Graph
from topologicpy.Dictionary import Dictionary
from topologicpy.Helper import Helper

import json
from rdflib import Graph as RDFGraph
from rdflib import URIRef, Literal, Namespace

# def convert_to_rdf_bot(graph): #Not working yet
#     # Create an RDF graph
#     rdf_graph = RDFGraph()
    
#     # Define namespaces
#     rdf = Namespace("http://www.w3.org/1999/02/22-rdf-syntax-ns#")
#     rdfs = Namespace("http://www.w3.org/2000/01/rdf-schema#")
#     bot = Namespace("https://w3id.org/bot#")
    
#     # Add vertices as RDF resources
#     for node, attributes in graph['nodes'].items():
#         node_uri = URIRef(bot[node])
#         rdf_graph.add((node_uri, rdf.type, bot["Vertex"]))
#         for key, value in attributes.items():
#             rdf_graph.add((node_uri, bot[key], Literal(value)))
    
#     # Add edges as RDF triples
#     for edge, attributes in graph['edges'].items():
#         source = attributes["source"]
#         target = attributes["target"]
#         source_uri = URIRef(bot[source])
#         target_uri = URIRef(bot[target])
#         rdf_graph.add((source_uri, bot["connectsTo"], target_uri))
#         for key, value in attributes.items():
#             rdf_graph.add((source_uri, bot[key], Literal(value)))
    
#     return rdf_graph


def convert_to_rdf(graph):
    # Create an RDF graph
    rdf_graph = RDFGraph()
    
    # Define namespaces
    rdf = Namespace("http://www.w3.org/1999/02/22-rdf-syntax-ns#")
    rdfs = Namespace("http://www.w3.org/2000/01/rdf-schema#")
    ex = Namespace("http://example.org/")
    
    # Add vertices as RDF resources
    for node, attributes in graph['nodes'].items():
        node_uri = URIRef(ex[node])
        rdf_graph.add((node_uri, rdf.type, ex["Vertex"]))
        for key, value in attributes.items():
            rdf_graph.add((node_uri, ex[key], Literal(value)))
    
    # Add edges as RDF triples
    for edge, attributes in graph['edges'].items():
        source = attributes["source"]
        target = attributes["target"]
        source_uri = URIRef(ex[source])
        target_uri = URIRef(ex[target])
        rdf_graph.add((source_uri, ex["connectsTo"], target_uri))
        for key, value in attributes.items():
            rdf_graph.add((source_uri, ex[key], Literal(value)))
    
    return rdf_graph



c = CellComplex.Prism()
c = Topology.Rotate(c, degree=32.5)
g = Graph.ByTopology(c, toExteriorTopologies=True)
json_data = Graph.JSONData(g)
rdf_graph = convert_to_rdf(json_data)
# Print RDF graph
print(rdf_graph.serialize(format="turtle"))
```
